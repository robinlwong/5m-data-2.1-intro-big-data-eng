
### Question 1

Question: From the `movies` collection, return the documents with the `plot` that starts with `"war"` in acending order of released date, print only title, plot and released fields. Limit the result to 5.


**Answer:**

```python
# ============================================================================
# Question 1
# ============================================================================
def war_movies(movies_collection):
    """
    Enter Your Solution Here
    """
    results = []
    
    # Start your solutions

    results = list(movies_collection.find(              # list is the wrapper
        {"plot": {"$regex": "^war", "$options": "i"}},  # Filter by plot starting with "war", case insensitive
        {"_id":0, "title":1, "plot":1, "released":1}    # !Display _id, display title, plot, released
    ).sort("released", 1).limit(5))                     # Sort by release date null values first, display 5
    
    # Your solution ends here
    
    return results
```

### Question 2

Question: Group by `rated` and count the number of movies in each.

**Answer:**

```python
# ============================================================================
# Question 2
# ============================================================================
def group_by_rated(movies_collection):
    """
    Group by rated and count the number of movies in each.
    """
    results = []
    
    # Start your solution

    pipeline = [                            # best break in the world
        {
            "$group": {
                "_id": "$rated",            # group documents by 'rated' field don't forget the underscore
                "movie_count": {"$sum":1}   # group +1 to movie_count for each title
            }
        }
    ]

    results = list(movies_collection.aggregate(pipeline))   # return results

    # End your solution
    
    return results
```


### Question 3

Question: Count the number of movies with 3 comments or more.


**Answer:**


```python
# ============================================================================
# Question 3
# ============================================================================
def count_movies_with_comments(movies_collection):
    """
    Count the number of movies with 3 comments or more.
    """
    count = 0
    
    # Start your solution

    query = {"num_mflix_comments": {"$gte": 3}}         # boundry case comments >=3
    count = movies_collection.count_documents(query)    # calculate internally then return int count of query
    # End your solution
    
    return count
```
